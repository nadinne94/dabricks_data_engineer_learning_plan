```sql
USE adventureworks.silver;

CREATE OR REPLACE TABLE gold.fact_sales_orders AS
SELECT
  o.sales_order_id,
  o.customer_id,

  o.bill_to_address_id,
  o.ship_to_address_id,
  
  o.order_date,
  o.ship_date,
  o.due_date,

  c.company_name,
  c.sales_person,

  o.sub_total,
  o.tax_amt,
  o.freight
  
FROM
  sales_orders o
JOIN
  customers c
ON
  o.customer_id = c.customer_id
```

```sql
SELECT *
FROM
  gold.fact_sales_orders
LIMIT 10;
```
