```sql
USE adventureworks.silver;

CREATE OR REPLACE TABLE gold.fact_sales_items AS
SELECT
  
  i.sales_order_id,
  i.sales_order_detail_id,
  i.product_id,

  p.product_name,
  p.standard_cost,
  p.list_price,
  p.category,
  p.model,
    
  i.order_qty AS quantity_sold,
  i.unit_price,
  i.line_total AS revenue

FROM 
  sales_items i
JOIN
  products p
ON
  i.product_id = p.product_id
```
