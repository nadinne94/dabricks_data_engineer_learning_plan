```sql
USE adventureworks.bronze;

CREATE OR REPLACE TABLE silver.sales_orders AS
SELECT
  
  CAST(sales_order_id AS INT) AS sales_order_id,
  CAST(customer_id AS INT) AS customer_id,
  CAST(bill_to_address_id AS INT),
  CAST(ship_to_address_id AS INT),
  
  online_order_flag,
  TO_DATE(order_date, 'yyyy-MM-dd') AS order_date,
  TO_DATE(ship_date,  'yyyy-MM-dd') AS ship_date,
  TO_DATE(due_date,   'yyyy-MM-dd') AS due_date,

  CAST(REPLACE(sub_total, ',', '.') AS DECIMAL(18,2)) AS sub_total,
  CAST(REPLACE(tax_amt, ',', '.') AS DECIMAL(18,2)) AS tax_amt,
  CAST(REPLACE(freight, ',', '.') AS DECIMAL(18,2)) AS freight,
  CAST(REPLACE(total_due, ',', '.') AS DECIMAL(18,2)) AS total_due


FROM sales_order_header
```

```sql
SELECT *
FROM silver.sales_orders
```
