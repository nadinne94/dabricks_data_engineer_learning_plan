```sql
USE adventureworks.bronze;
drop table if exists sales_items;

CREATE OR REPLACE TABLE silver.sales_items AS
SELECT
    CAST(h.sales_order_id AS INT) AS sales_order_id,
    CAST(d.sales_order_detail_id AS INT) AS sales_order_detail_id,
    CAST(h.customer_id AS INT) AS customer_id,
    CAST(d.product_id AS INT) AS product_id,

    CAST(d.order_qty AS INT) AS order_qty,
    CAST(REPLACE(d.unit_price, ',', '.') AS DECIMAL(18,2)) AS unit_price,
    CAST(REPLACE(d.unit_price_discount, ',', '.') AS DECIMAL(5,2)) AS unit_price_discount,
    CAST(REPLACE(d.line_total, ',', '.') AS DECIMAL(18,2)) AS line_total,

    h.status
FROM
  sales_order_detail d
JOIN
  sales_order_header h
ON
  d.sales_order_id = h.sales_order_id
```

```sql
SELECT * FROM adventureworks.silver.sales_items
LIMIT 10
```
