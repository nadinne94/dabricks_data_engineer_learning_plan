```sql
use adventureworks.bronze;

CREATE OR REPLACE TABLE silver.customers_address AS
SELECT
  CAST(ca.customer_id AS INT),
  CAST(ca.address_id AS INT),
  a.address_line1 AS address,
  a.city,
  a.state_province,
  a.country_region,
  a.postal_code
FROM customer_address ca
INNER JOIN address a
ON ca.address_id = a.address_id
```

```sql
%python
df = spark.sql("""SELECT * FROM silver.customers_address""")
#checar duplicados
duplicate_existis = df.count() > df.dropDuplicates().count()
print(f"Duplicate existis: {duplicate_existis}")

#Boolean para taskValues
dbutils.jobs.taskValues.set(key="has_duplicates", value=duplicate_existis)
```

```sql
%python

# checar nulos
null_count = df.filter(
    df.customer_id.isNull() |
    df.first_name.isNull() |
    df.last_name.isNull() |
    df.phone.isNull() |
    df.email_address.isNull() |
    df.company_name.isNull() |
    df.sales_person.isNull()
).count()
null_existis = null_count > 0
print(f"Null existis: {null_existis}")

#Boolean para taskValues
dbutils.jobs.taskValues.set(key="null_existis",value=null_existis)
```
