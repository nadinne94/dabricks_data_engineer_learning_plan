###CustomerAddress

```sql
select address_id, count(*)
from customer_address
group by address_id
order by count(*) desc
```

###Product_description

```sql
SELECT *
FROM adventureworks.bronze.product_description
```

```sql
SELECT
  SUM(CASE WHEN description LIKE '%?%' THEN 1 ELSE 0 END) AS com_interrogacao,
  SUM(CASE WHEN description NOT LIKE '%?%' OR description IS NULL THEN 1 ELSE 0 END) AS sem_interrogacao
FROM adventureworks.bronze.product_description;
```

###SalesOrdeHeader

```sql
select count(*) as total from adventureworks.bronze.sales_order_header
```

```sql
SELECT
  SUM(CASE WHEN ship_to_address_id = bill_to_address_id THEN 1 ELSE 0 END) AS same_address,
  SUM(CASE WHEN ship_to_address_id <> bill_to_address_id THEN 1 ELSE 0 END) AS diferent_address
FROM adventureworks.bronze.sales_order_header;
```

```sql
describe table adventureworks.bronze.sales_order_header
```

```sql
SELECT
  SUM(CASE WHEN online_order_flag LIKE 'True' THEN 1 ELSE 0 END) AS online,
  SUM(CASE WHEN online_order_flag LIKE 'False' THEN 1 ELSE 0 END) AS not_online
FROM adventureworks.bronze.sales_order_header;
```

###SalerOrderDetail

```sql
describe table adventureworks.bronze.sales_order_detail
```

```sql
SELECT
  SUM(CASE WHEN unit_price_discount = '0' THEN 1 ELSE 0 END) AS no_discount,
  SUM(CASE WHEN unit_price_discount <> '0' THEN 1 ELSE 0 END) AS discount
FROM adventureworks.bronze.sales_order_detail;
```
