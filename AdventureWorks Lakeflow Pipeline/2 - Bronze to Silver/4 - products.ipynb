```sql
USE adventureworks.bronze;

CREATE OR REPLACE TABLE silver.products AS
SELECT
    CAST(p.product_id AS INT),
    p.name AS product_name,
    CAST(REPLACE(p.standard_cost, ',', '.') AS DECIMAL(18,2)) AS standard_cost,
    CAST(REPLACE(p.list_price, ',', '.') AS DECIMAL(18,2)) AS list_price,
    pc.name  AS category,
    pm.name AS model
FROM product p
LEFT JOIN product_category pc
    ON p.product_category_id = pc.product_category_id
LEFT JOIN product_model pm
    ON p.product_model_id = pm.product_model_id
LEFT JOIN product_model_product_description pmpd
    ON pm.product_model_id = pmpd.product_model_id
    
```

```sql
select * from adventureworks.silver.products
limit 10;
```
