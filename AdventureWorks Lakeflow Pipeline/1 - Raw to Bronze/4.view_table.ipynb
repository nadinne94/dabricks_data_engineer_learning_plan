```python
schema = 'adventureworks.bronze'

# List tables in the schema using Spark catalog
# Only include tables (not views)
tables = [t.name for t in spark.catalog.listTables(schema) if t.tableType == 'MANAGED' or t.tableType == 'EXTERNAL']
```

```python
for table in tables:
    df = spark.sql(f"SELECT * FROM {schema}.{table}")
    print(f"Table: {table}")
    display(df)
    print("")
```
