```python
from pyspark.sql import functions as F
import re
```

```python
def to_snake_case(name: str) -> str:
    name = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', name)
    name = re.sub('([a-z0-9])([A-Z])', r'\1_\2', name)
    return name.lower()
```

```python
catalog = dbutils.widgets.get("catalog")
schema = dbutils.widgets.get("schema")


tables = [t.name for t in spark.catalog.listTables(f"{catalog}.{schema}")]
```

```python
for t in tables:
    df = spark.read.table(f"{catalog}.{schema}.{t}")
    df = df.toDF(*[to_snake_case(c) for c in df.columns])
    df.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable(f"{catalog}.{schema}.{t}")
```
