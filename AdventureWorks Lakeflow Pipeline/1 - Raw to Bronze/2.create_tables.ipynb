```python
import re

def to_snake_case(name: str) -> str:
    name = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', name)
    name = re.sub('([a-z0-9])([A-Z])', r'\1_\2', name)
    return name.lower()
```

```python
source_path = f'/Volumes/adventureworks/dataset/raw/sales/'

tables = dbutils.widgets.get('tables')

my_catalog = dbutils.widgets.get('catalog')
my_schema = dbutils.widgets.get('schema')

table_name = to_snake_case(tables)
```

```python
df = (spark
        .read
        .format("csv")
        .option("header", "true")
        .option("sep", ";")
        .option("encoding", "ISO-8859-1")
        .load(f'{source_path}/{tables}.csv')
    )
    
df.write.mode('overwrite').saveAsTable(f'{my_catalog}.{my_schema}.{table_name}')
```
